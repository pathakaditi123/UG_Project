```python
df=pd.read_csv('/content/drive/MyDrive/Training_Data.csv')
```

```python
df
```

```python
X=df.iloc[:,1:4]
```

```python
X
```

```python
Y=df.iloc[:,0]
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
len(Y)
```

```python
Y=np.array(Y)
Y=Y.reshape(-1,1)
```

```python
scaler = MinMaxScaler()
X=scaler.fit_transform(X)

# Y=scaler.fit_transform(Y)
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test = train_test_split(X,Y ,
                                   random_state=104, 
                                   test_size=0.25, 
                                   shuffle=True)
```

```python
Y_test[0]
```

```python
hidden_units1 = 576
hidden_units2 = 256
hidden_units3 = 128
hidden_units4 = 64

learning_rate = 0.005
def build_model_using_sequential():
  model = Sequential([
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.2),
    Dense(hidden_units2, kernel_initializer='normal', activation='relu'),
    Dropout(0.2),
    Dense(hidden_units3, kernel_initializer='normal', activation='relu'),
    Dropout(0.2),
    Dense(hidden_units4, kernel_initializer='normal', activation='relu'),
    Dense(1, kernel_initializer='normal', activation='linear')
  ])
  return model
# build the model
model = build_model_using_sequential()
```

```python
msle = MeanSquaredLogarithmicError()
model.compile(
    loss=msle, 
    optimizer=Adam(learning_rate=learning_rate), 
    metrics=[msle]
)
```

```python
history = model.fit(
    X_train, 
    Y_train, 
    epochs=100, 
    batch_size=64,
    validation_split=0.2
)
```

```python
def plot_history(history, key):
  plt.plot(history.history[key])
  plt.plot(history.history['val_'+key])
  plt.xlabel("Epochs")
  plt.ylabel(key)
  plt.legend([key, 'val_'+key])
  plt.show()

  
plot_history(history, 'mean_squared_logarithmic_error')
```

```python
Y_pred=model.predict(X_test)
```

```python
Y_pred
```

```python
Y_test
```

```python
sum_error=0
errors=[]
for i in range(len(Y_pred)):
  per=abs(Y_pred[i]-Y_test[i])/max(Y_test[i],Y_pred[i])
  sum_error+=per
  errors.append(per*100)
  # print(per*100)

print(sum_error/len(Y_pred)*100)
errors.sort()
```

**Thus we are getting 10% error,thus model is 90% correct.**

```python
errors
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.scatter(Y_test,Y_pred, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(Y_test), max(Y_pred))
p2 = min(min(Y_pred), min(Y_test))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```
